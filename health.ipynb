```python
h = health.copy()
```

```python
h = h.rename(columns={"Country Name": "Country_Name","Country Code": "Country_Code", "Series Name": "Series_Name", "Series Code": "Series_Code"})
```

```python
h.Series_Name.unique()
```

```python
h.columns = h.columns.str.replace(' \[YR\d+\]', '', regex=True)
```

```python
h_china = h.loc[h["Country_Name"] == "China", :]
```

```python
h_china_1995_2018 = h_china.drop(h_china.loc[:,'1970':'1994'].columns, axis=1)
```

```python
h_china_1995_2018=h_china_1995_2018.drop(['Country_Code','Series_Code'], axis=1) 
```

```python
result = pd.concat([top10depart_China_1995_2018, h_china_1995_2018], axis=0)
```
